```python
#positional argument
def Calculation(a, b):
  print("Addition: ", a+b)
  print("Subtraction: ", a-b)
  print("Multiplication: ", a*b)
  print("Division: ", a/b)
  return "Calculation is complete"
```

```python
#default argument
def area_perimeter(width=4,height=8):
  area = width*height
  perimeter = 2*(width+height)
  return "Area of rectangle is", area, "and perimeter is",perimeter
```

```python
print(area_perimeter(10))
```

```python
#bar plot--it work on numerical as well as categorial data
import matplotlib.pyplot as plt
import numpy as np
fruit = np.array(["Apple","Banana","Orange","Watermelon"])
quantity = np.array([100,120,80,70])
color = np.array(["red","yellow","orange","green"])
plt.bar(fruit,quantity,color = color)
plt.xlabel("Fruit")
plt.ylabel("Quantity of fruit")
plt.title("Fruit vs fruit quantity")
plt.show()
```

```python
fruit = np.array(["Apple","Banana","Orange","Watermelon"])
quantity = np.array([100,120,80,70])
color = np.array(["red","yellow","orange","green"])
plt.barh(fruit,quantity,color = color)
plt.xlabel("Fruit")
plt.ylabel("Quantity of fruit")
plt.title("Fruit vs fruit quantity")
plt.show()
```

```python
#pie plot
percentage_of_fruit = np.array([20,25,30,25])
fruit = np.array(["Apple","Banana","Orange","Watermelon"])
myexplode = np.array([0.2,0,0,0])
plt.pie(percentage_of_fruit,labels = fruit,explode = myexplode)
plt.legend(title = "Fruit",loc = "best")
plt.show()
```

```python
age_of_car = np.array([4,5,6,7,8,9,10,12,14])
speed_of_car = np.array([150,140,120,110,100,90,80,70,60])
plt.scatter(age_of_car,speed_of_car)
plt.xlabel("Age of car")
plt.ylabel("Speed of car")
plt.title("Age of car vs speed of car")
plt.show()
```
